# Tipos de variáveis quantitativas e qualitativas:

| Tipo de Variável  | Subtipo               | Descrição                                                                                           | Exemplos                        |
|-------------------|-----------------------|-----------------------------------------------------------------------------------------------------|---------------------------------|
| **Quantitativa**  | **Discreta**           | Valores numéricos inteiros. Não podem assumir valores intermediários entre inteiros.                 | Número de filhos, carros, etc.  |
|                   | **Contínua**           | Pode assumir qualquer valor dentro de um intervalo contínuo, incluindo frações e decimais.           | Altura, peso, tempo, temperatura|
| **Qualitativa**   | **Nominal**            | Categorização sem uma ordem intrínseca.                                                              | Cor dos olhos, estado civil     |
|                   | **Ordinal**            | Categorização com uma ordem ou hierarquia, mas sem uma diferença numérica exata entre categorias.     | Grau de escolaridade, níveis de satisfação (baixa, média, alta)  |


# Explicação:
- Quantitativa: Representa medidas ou contagens numéricas.

  - Discreta: São contáveis e não admitem frações.
  - Contínua: Podem assumir valores em um intervalo, incluindo frações e decimais.
  - 
- Qualitativa: Representa categorias ou atributos.

  - Nominal: Não tem ordem ou hierarquia.
  - Ordinal: Tem uma ordem, mas as diferenças entre os níveis não são necessariamente mensuráveis.-

# Criar o gráfico de barras horizontais com Seaborn

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregar os dados e remover linhas com valores faltantes nas colunas 'Price (in rupees)' e 'Bathroom'
clean_data = pd.read_csv('datasets/house_prices.csv')[['Price (in rupees)', 'Bathroom']].dropna()

# Converter os preços de rúpias para reais (considerando a taxa de câmbio de 5)
clean_data['Price (in R$)'] = clean_data['Price (in rupees)'] / 5

# Agrupar por número de banheiros e calcular o preço médio em reais
mean_prices = clean_data.groupby('Bathroom')['Price (in R$)'].mean().reset_index()

# Ordenar os dados pelo número de banheiros
mean_prices = mean_prices.sort_values(by='Bathroom')

# Criar o gráfico de barras horizontais
plt.figure(figsize=(8, 6))
sns.barplot(x='Price (in R$)', y='Bathroom', data=mean_prices, palette="Blues_d")

# Adicionar títulos e rótulos
plt.title('Média de Preço por Número de Banheiros', fontsize=14)
plt.xlabel('Preço Médio (em R$)', fontsize=12)
plt.ylabel('Número de Banheiros', fontsize=12)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
